```python
def remove_emoji(string):
    emoji_pattern = re.compile(
        "["
         u"\U0001F600-\U0001F64F"  # emoticons
         u"\U0001F300-\U0001F5FF"  # symbols & pictographs
         u"\U0001F680-\U0001F6FF"  # transport & map symbols
         u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
         u"\U00002702-\U000027B0"
         u"\U000024C2-\U0001F251"
         "]+",
         flags=re.UNICODE,
    )
    return emoji_pattern.sub(r"", string)
    
```

```python
for i in range(len(tweet_data)):
    new_text = remove_emoji(str(tweet_data['text'][i]))
    tweet_data['text'][i] = new_text
```

```python
import gensim
from gensim.models import Word2Vec, KeyedVectors
```

```python
arr = []
for i in range(len(data)):
    arr.append(data['text'][i])
```

```python
arr[0]
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words = []
for sent in arr:
    sent_token = sent_tokenize(str(sent))
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
# training gensim model
model = gensim.models.Word2Vec(words, vector_size = 250, epochs = 50)
```

```python
model.corpus_count
```

```python
import numpy as np
def avg_word2vec(doc):
    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key ],axis=0)
```

```python
from tqdm import tqdm
```

```python
# applying word2vec on all sentences
x = []
for i in tqdm(range(len(words))):
    x.append(avg_word2vec(words[i]))
```

```python
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1], 1)
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1], 1)
```

```python
model = Sequential()
model.add(LSTM(60,return_sequences=True, dropout=0.25, recurrent_dropout=0.1, input_shape = (5,1)))
model.add(Bidirectional(LSTM(25,return_sequences=False, dropout=0.25, recurrent_dropout=0.1)))
model.add(Dense(3, activation = 'softmax'))
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train,y_train, validation_data=(x_test,y_test), epochs=50, batch_size=50, verbose=1)
```
